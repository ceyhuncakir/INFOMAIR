```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Set up the figure and axes for subplots
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

# Plot count of restaurants by pricerange
df['pricerange'].value_counts().plot(kind='bar', ax=axs[0])
axs[0].set_title('Count of Restaurants by Price Range')
axs[0].set_xlabel('Price Range')
axs[0].set_ylabel('Count')

# Plot count of restaurants by area
df['area'].value_counts().plot(kind='bar', ax=axs[1])
axs[1].set_title('Count of Restaurants by Area')
axs[1].set_xlabel('Area')
axs[1].set_ylabel('Count')

# Plot count of restaurants by food type
df['food'].value_counts().plot(kind='bar', ax=axs[2])
axs[2].set_title('Count of Restaurants by Food Type')
axs[2].set_xlabel('Food Type')
axs[2].set_ylabel('Count')

# Adjust layout
plt.tight_layout()
plt.show()
```
